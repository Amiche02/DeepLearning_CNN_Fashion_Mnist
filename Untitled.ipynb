```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_val = x_val.reshape(x_val.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
x_train.shape, x_test.shape, x_val.shape
```

## CNN

```python
cnn_model = Sequential([
    Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=2),
    Dropout(0.25),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
    Conv2D(128, (3,3), activation='relu'),
    Dropout(0.4),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])
```

```python
cnn_model.summary()
```

```python
!pip install pydot
plot_model(cnn_model, to_file='model.png', dpi=50)
```

```python
cnn_model.compile(loss=keras.losses.SparseCategoricalCrossentropy(),
                 optimizer = Adam(learning_rate=0.001),
                 metrics=["accuracy"])

history = cnn_model.fit(x_train,
                       y_train,
                       batch_size=4096,  
                       epochs=100,
                       validation_data=(x_val, y_val))
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Training - Loss Functio')

plt.subplot(2,2,2)
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title('Train - Accuracy')
```

```python
cnn_model.evaluate(x_test, y_test)
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

plt.subplot(2,1,1)
plt.plot(epochs, accuracy, label='Training Accuracy')
plt.plot(epochs, val_accuracy, label='Validation Accuracy')
plt.title('Training and validation accuracy')
plt.legend()

plt.subplot(2,1,2)
plt.plot(epochs, loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.tight_layout()
```

```python
predicted_classes = np.argmax(cnn_model.predict(x_test), axis=-1)

correct = np.nonzero(predicted_classes == y_test)
incorrect = np.nonzero(predicted_classes != y_test)

target_name = [f"Class {i for i in range(10)}"]
print(classification_report(y_test, predicted_classes))
```

```python
l = 5
w = 5

fig, ax = plt.subplots(l, w, figsize=(12, 12))
ax = ax.ravel()

for i in np.arange(0, l*w):
    ax[i].imshow(x_test[i].reshape(28,28))
    ax[i].set_title(f"Prediction Class = {predicted_classes[i]:0.1f}\n Original Class = {y_test[i]:0.1f}")
    ax[i].axis('off')
    
plt.subplots_adjust(wspace=0.5)
```

```python
print(f"Correct predicted classes : {len(correct[0])}")
print(f"Incorrect predicted classes : {len(incorrect[0])}")
```

```python
def plot_images(data_index,cmap="Blues"):
    # Plot the sample images now
    f, ax = plt.subplots(4,4, figsize=(15,15))

    for i, indx in enumerate(data_index[:16]):
        ax[i//4, i%4].imshow(x_test[indx].reshape(28,28), cmap=cmap)
        ax[i//4, i%4].axis('off')
        ax[i//4, i%4].set_title("True:{}  Pred:{}".format(class_names[y_test[indx]],class_names[predicted_classes[indx]]))
    plt.show() 
```
